```python
from PIL import Image, ImageSequence

def tiff_to_png_layer(tiff_path, png_path):
    """
    Opens a TIFF file, extracts the last layer, and saves it as a PNG.
    """
    try:
        # Open the TIFF file
        with Image.open(tiff_path) as img:
            # Iterate through all frames (layers) in the TIFF
            for i, page in enumerate(ImageSequence.Iterator(img)):
                # Save the last page as a PNG
                if i == img.n_frames - 1:
                    page.save(png_path, 'PNG')
                    print(f"Last layer saved to {png_path}")
    except FileNotFoundError:
        print(f"Error: The file {tiff_path} was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage:
tiff_to_png_layer("test_1.1.tif", "out_1.1.png")
```
