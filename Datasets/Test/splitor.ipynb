```python
import os
from PIL import Image

def resize_to_512(img):
    """Resize any image to 512x512."""
    return img.resize((512, 512), Image.LANCZOS)

def split_image(img, n):
    """Split a PIL image into n x n tiles."""
    tiles = []
    w, h = img.size
    tile_w = w // n
    tile_h = h // n

    for row in range(n):
        for col in range(n):
            left = col * tile_w
            upper = row * tile_h
            right = left + tile_w
            lower = upper + tile_h
            tile = img.crop((left, upper, right, lower))
            tiles.append(tile)
    return tiles

def process_folder(input_folder, output_folder, n=2):
    """Process all PNGs: resize  split  save tiles."""
    os.makedirs(output_folder, exist_ok=True)

    for filename in os.listdir(input_folder):
        if not filename.lower().endswith(".png"):
            continue

        input_path = os.path.join(input_folder, filename)
        base_name = os.path.splitext(filename)[0]

        print(f"Processing: {filename}")

        # Load + resize
        img = Image.open(input_path)
        img_resized = resize_to_512(img)

        # Split
        tiles = split_image(img_resized, n)

        # Save tiles with numbering
        for i, tile in enumerate(tiles, start=1):
            index = f"{i:02d}"  # 01, 02, 03, ...
            output_name = f"{base_name}_{index}.png"
            output_path = os.path.join(output_folder, output_name)
            tile.save(output_path)

        print(f"Saved {len(tiles)} tiles from {filename}")

    print("Done!")

# -----------------------------
# Example Usage
# -----------------------------
if __name__ == "__main__":
    n=2
    process_folder(
        input_folder="../pngs_max/OSD-627/version-2/microscopy",
        output_folder="../pngs_max_split/OSD-627/version-2/microscopy",
        n=n  # change to 3, 4, etc.
    )

    process_folder(
        input_folder="../pngs_max/OSD-627/version-1/microscopy",
        output_folder="../pngs_max_split/OSD-627/version-1/microscopy",
        n=n  # change to 3, 4, etc.
    )
    
    process_folder(
        input_folder="../pngs_mean/OSD-627/version-2/microscopy",
        output_folder="../pngs_mean_split/OSD-627/version-2/microscopy",
        n=n  # change to 3, 4, etc.
    )

    process_folder(
        input_folder="../pngs_mean/OSD-627/version-1/microscopy",
        output_folder="../pngs_mean_split/OSD-627/version-1/microscopy",
        n=n  # change to 3, 4, etc.
    )



```
